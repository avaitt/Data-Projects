```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb 

%matplotlib inline 
```

```python
fuel_econ = pd.read_csv(r'\Users\Acer\Desktop\fuel_econ.csv')
print(fuel_econ.shape) # print rows, columns
fuel_econ.head(6)
```

```python
sedan_classes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 
                 'Midsize Cars', 'Large Cars']
vclasses = pd.api.types.CategoricalDtype(ordered = True, categories = sedan_classes)
# Type for categorical data with the categories and orderedness
fuel_econ['VClass'] = fuel_econ['VClass'].astype(vclasses);


```

Uses panda's astype function to convert the vlcass column from a plain object type into an ordered categorical type putting the classes in order of size. This way the sorting of levels is automatic

```python
sb.violinplot(data = fuel_econ, x = 'VClass', y = 'comb');
```

```python
base_color = sb.color_palette()[0]
# makes sure each level is the same color
sb.violinplot(data = fuel_econ, x = 'VClass', y = 'comb',
              color = base_color, inner = None);
# inner = None parameter clears up the middle of the plots
plt.xticks(rotation = 15);
# slightly rotates the labels
```

Compact and Midsize Cars have the best efficiencies, but the distrubution of the Midsize Cars is broader than the Compact

```python
base_color = sb.color_palette()[0]
# makes sure each level is the same color
sb.violinplot(data = fuel_econ, x = 'VClass', y = 'comb',
              color = base_color);
plt.xticks(rotation = 15);
# slightly rotates the labels
```

Inside each curve, there is a black shape with a white dot inside. This is the miniature box plot (another quantitative vs qualitative plot type that takes a summary approach to displaying the data)
