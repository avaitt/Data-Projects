```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb 

%matplotlib inline 
```

```python
fuel_econ = pd.read_csv(r'\Users\Acer\Desktop\fuel_econ.csv')
print(fuel_econ.shape) # print rows, columns
fuel_econ.head(6)
```

```python
sb.regplot(data = fuel_econ, x = 'displ', y = 'comb', x_jitter = 0.04,
           scatter_kws = {'alpha' : 1/10}, fit_reg = False );
# added jitter and transparency to see where the main bulk of the data lies
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

```python
plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb');
plt.colorbar() 
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

Similar to a histogram, the plotting area is divided into a grid and the number of points in each grid rectangle is added up. Since there won't be room for bar heights, counts are indicated instead by grid cell color. A heat map can be implemented with Matplotlib's hist2d function.

We add a colorbar function call to add a colorbar to the side of the plot, showing the mapping from counts to colors.

```python
plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb', cmin = 0.5);
plt.colorbar() 
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

The "cmin" parameter specifies the minimum value in a cell before it will be plotted. By adding a cmin = 0.5 parameter to the hist2d call, this means that a cell will only get colored if it contains at least one point.

```python
plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb', cmin = 0.5,
           cmap = 'viridis_r');
# reverse the default "viridis" color palette, by setting cmap = 'viridis_r'
plt.colorbar() 
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

```python
fuel_econ[['displ', 'comb']].describe()
# look at descriptive statistics for 2 variables to decide bin sizes for x and y
# bin size of 0.3 for displacement ("step")
# bin size of 3 for fuel efficiency ("step")
```

```python
bins_x = np.arange(0.6, 7+0.3, 0.3) # start, stop, step
bins_y = np.arange(12, 58+3, 3)
plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb', cmin = 0.5,
           cmap = 'viridis_r', bins = [bins_x, bins_y]);
plt.colorbar() 
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```
